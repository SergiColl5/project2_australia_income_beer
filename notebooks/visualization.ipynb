```python
'''With the given data, you can prepare the following plots to visualize the relationships and patterns in the data:

Line plot: To see the trend of median income over the years.

Bar plot: To compare the number of sponsored visas between different economic activities.

Scatter plot: To see the relationship between median income and the number of sponsored visas for each economic activity.

Grouped bar plot: To compare the median income and number of sponsored visas between different economic activities.

Box plot: To see the distribution of median income for each economic activity.

Violin plot: To see the distribution of both median income and number of sponsored visas for each economic activity.

Heatmap: To see the relationship between year, median income, and number of sponsored visas.

These are just a few examples. The best plots to prepare will depend on what you want to explore and communicate with the data.'''
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
all_together = pd.read_excel('../data/all_together.xlsx')
all_together.columns
```

```python


# Group the data by year and calculate the mean median income
mean_median_income_by_year = all_together.groupby('Year')['Median_income'].mean().reset_index()

# Plot the mean median income over the years
plt.plot(mean_median_income_by_year['Year'], mean_median_income_by_year['Median_income'])

# Add labels and title to the plot
plt.xlabel('Year')
plt.ylabel('Mean Median Income')
plt.title('Mean Median Income over the Years')

# Show the plot
plt.show()
```

```python


# Group the data by year and calculate the mean number of sponsored visas
mean_sponsored_visas_by_year = all_together.groupby('Year')['Sponsored_visas'].mean().reset_index()

# Plot the mean number of sponsored visas over the years
plt.plot(mean_sponsored_visas_by_year['Year'], mean_sponsored_visas_by_year['Sponsored_visas'])

# Add labels and title to the plot
plt.xlabel('Year')
plt.ylabel('Mean Sponsored Visas')
plt.title('Mean Sponsored Visas over the Years')

# Show the plot
plt.show()
```
